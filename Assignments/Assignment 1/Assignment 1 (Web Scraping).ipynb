## 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’

```python
# Load necessary Libaries
import requests
from bs4 import BeautifulSoup as bs

# Load the webpage content
r = requests.get('https://en.wikipedia.org/wiki/Main_Page')

# Convert into Beautiful Soup object
Wikipedia = bs(r.content)

# print the webpage
print(Wikipedia.prettify())
```

```python
# Using the find_all method to find all the available headers on Title page
titles = Wikipedia.find_all(['h1', 'h2', 'h3'])

# Print the list of Headers
print('The List of Headers on the Title Page of Wikipedia is as follows :', *titles, sep = '\n\n')
```

## 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
r1 = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
IMDb = bs(r1.content)
print(IMDb.prettify())
```

```python
table = IMDb.find('tbody', attrs = {'class' : 'lister-list'}).find_all('tr')
print(table)
```

```python
Name = []
IMDb_Rating = []
Year_of_Release = []

for data in table:
    if data == 100 :
        break
    Name.append(data.find('td', attrs = {'class' : 'titleColumn'}).find('a').get_text())
    IMDb_Rating.append(float(data.find('td', attrs = {'class' : 'ratingColumn imdbRating'}).find('strong').get_text()))
    Year_of_Release.append(int(data.find('td', attrs = {'class' : 'titleColumn'}).find('span').text[1:5]))
```

## 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

## 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
r4 = requests.get('https://bookpage.com/reviews')
book_reviews = bs(r4.content)
print(book_reviews.prettify)
```

```python
book_details = book_reviews.find('div', attrs = {'class' : 'span9'}).find_all('div', attrs = {'class' : 'flex-article-content'})
book_details
```

```python
Book_Name = []
Author_Name = []
Genre = []
Review = []

for data2 in book_details :
    Author_Name.append(str(data2.find('p', attrs = {'class' : 'sans bold'}).text).strip())
    Book_Name.append(str(data2.find('h4').text).strip())
    Genre.append(str(data2.find('p', attrs = {'class' : 'genre-links hidden-phone'}).get_text()).strip())
    Review.append(data2.find('div', attrs = {'class' : 'read-full'}).find('a').get('href'))
```

```python
Output_4 = list(zip(Book_Name, Author_Name, Genre, Review))
Output_4
```

```python
df_Output_4 = pd.DataFrame(Output_4, columns = ['Book_Name', 'Author_Name', 'Genre', 'Review'])
df_Output_4.index = df_Output_4.index + 1 
df_Output_4[:5]
```
