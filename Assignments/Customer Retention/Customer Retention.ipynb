```python
# Importing the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the excel dataset
data_v1 = pd.read_excel('customer_retention_dataset.xlsx')
data_v1
```

```python
data_v1.info()
```

```python
# Printing the column name of all the available columns
for col in data_v1.columns :
    print(col)
```

```python
#Plotting the countplot for various variables

fig = plt.figure(figsize = (26,12))

ax1 = fig.add_subplot(2, 2, 1)
sns.countplot('5 Since How Long You are Shopping Online ?', data = data_v1)
plt.title('Ageing of individuals from the time they started shopping online')
plt.xlabel('Ageing Group')
plt.ylabel('Number of individuals')

ax2 = fig.add_subplot(2, 2, 2)
sns.countplot('7 How do you access the internet while shopping on-line?', data = data_v1)
plt.title('Distribution of individuals accessing internet for online shopping')
plt.xlabel('Various Medium of internet access for shopping online')
plt.ylabel('Number of individuals')

ax3 = fig.add_subplot(2, 2, 3)
sns.countplot('32 Shopping online is convenient and flexible', data = data_v1)
plt.title('Distribution of individuals on convenience and flexibility of Online Shopping')
plt.xlabel('Level of satisfaction')
plt.ylabel('Number of individuals')

ax4 = fig.add_subplot(2, 2, 4)
sns.countplot('6 How many times you have made an online purchase in the past 1 year?', data = data_v1)
plt.title('Number of times individual had done the shopping online')
plt.xlabel('Number of Times')
plt.ylabel('Number of individuals')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```

- Majority of customer uses mobile internet for shopping online
- Most of the customers started shopping way back more than 4 years, which is follows by customers who are following online shopping from 2 - 3 years 
- Majority of customers "strongly agree" and "agree" regarding the convenience and flexibility of online shopping
- During the past one year, majority of customers have shopped for less than 10 times - A detailed research on this, will help to understand the commodity
